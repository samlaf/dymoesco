```python
def generate_smooth_control(umin, umax, n):
    umin = np.array(umin, ndmin=1); umax = np.array(umax, ndmin=1);
    u_rough = np.random.uniform(umin, umax, (len(umin), n))
    sos = signal.butter(10, 0.1, output='sos')
    u = signal.sosfilt(sos, u_rough)
    return u
```
