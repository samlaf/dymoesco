# We show the basic functionalities of the single integrator model

```python
from dymoesco.dynamics.singleintegrator import SingleIntegrator
from dymoesco.control.misc_controllers import generate_smooth_control
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%load_ext autoreload
%autoreload 2
```

```python
# Parameters
t_span = (0,10)
dt = 0.1
t = np.arange(*t_span, dt)

umax = np.array([2,2])
umin = -umax
```

```python
# Single integrator plant
si = SingleIntegrator(dim=2)
# random control sequence
u = generate_smooth_control(umin, umax, t_span, dt)
# Generate trajectory
traj = si.simulate(u, t_span, [0,0], np.arange(*t_span,dt))
```

```python
%matplotlib inline
fig, (ax1, ax2) = plt.subplots(1,2)
si.plotu(traj, ax=ax1)
si.plotx(traj, ax=ax2)
```

```python
si.plot_phase(traj)
```

```python
%matplotlib notebook
si.animate(traj, tail_len=3)
```
